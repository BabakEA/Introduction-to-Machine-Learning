```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pylab as plt
```

```python
x=[i for i in range (-5,5)]
linear_y=[2*(i)+1 for i in x]
non_linear_y=[np.sin(i) for i in x]

df=pd.DataFrame(index=x,columns=["linear_data","nonlinear_data"])
df["linear_data"]=linear_y
df["nonlinear_data"]=non_linear_y
df.plot()
```

```python

plt.figure(figsize=(10, 4),dpi=70)
for col in df.columns:
    df[col].plot(label=col )

plt.grid()
plt.legend()
plt.title("Linear data vs nonlinear ")
plt.ylabel("Data value ") 
```
